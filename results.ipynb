```python
import matplotlib.pyplot as plt
import csv
from collections import defaultdict

midpoints = []
user_inputs = []
labels = []

with open("results.csv", newline="") as f:
    reader = csv.DictReader(f)
    for i, row in enumerate(reader):
        print(row)
        glyph = row["glyph_type"]
        A = float(row["sizeA"])
        B = float(row["sizeB"])
        C = float(row["sizeC"])

        midpoint = (A + C) / 2

        labels.append(f"{glyph}-{i+1}")
        midpoints.append(midpoint)
        user_inputs.append(B)

x = range(len(midpoints))

plt.figure(figsize=(12, 6))
bar_width = 0.2

plt.bar([i - bar_width/2 for i in x], midpoints, width=bar_width, color="green", label="Accurate midpoint")

plt.bar([i + bar_width/2 for i in x], user_inputs, width=bar_width, color="red", label="User input")

plt.xticks(x, labels, rotation=45, ha="right")
plt.ylabel("Size of glyphs")
plt.title("Comparing user inputs with accurate midpoints")
plt.legend()
plt.tight_layout()
plt.show()
```
